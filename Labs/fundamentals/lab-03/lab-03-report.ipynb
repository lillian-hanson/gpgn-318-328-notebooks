```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Pugh-Hanson-Data.csv")
data
```

```python
fig, axes=plt.subplots(1, 5, figsize=(20, 8), sharey=True, constrained_layout=True)

# Column names (as strings)
columns = ['1', '2', '3', '4', '5']

for ax, col in zip(axes, columns):
    ax.hist(data[col], bins=16, color='skyblue', edgecolor='black')
    ax.set_title(f'Flag {col}')
    ax.grid(color='lightgray', alpha=0.7)
    ax.set_xlim(0.001, 0.017)

fig.supxlabel("Self-Potential (V)")
fig.supylabel("Number of Occurences")

plt.suptitle('Self Potential Measurements at Varied Locations', fontsize=16)
plt.show()
```

In the figure above, Flag 1, 2, 4, and 5 show a right skew. Flag 3 shows fairly symmetric distribution with a slight right skew. 

The distributions shown on this histogram solidify the understanding I came to in my lab notes reflection, that there can be an incredible amount of sources of error in a measurement, which is why repeated measurements are so important. A significant source of error for this Lab may have been soil saturation, as it was raining at the beginning of the lab but the sun became very intense and drying througout the duration of the Lab. There was also the potential of an electrode contacting the flag or some other subsurface object, which could be hard to detect from aboveground but would give a different value for self-potential than the ground would. I would expect the sources of error in this lab to increase the standard deviation, reflecting the further spread of the data due to the error. 

```python
for col in data.select_dtypes(include="number").columns:
    n = data[col].count()
    mean_val = data[col].mean()
    std_val = data[col].std()

    # Standard deviation of the mean (standard error of the mean)
    std_mean = std_val / (n**0.5)

    print(f"Column: {col}")
    print(f"  Number of samples: {n}")
    print(f"  Mean: {mean_val:.6f}")
    print(f"  Standard deviation: {std_val:.6f}")
    print(f"  Standard deviation of the mean: {std_mean:.6f}\n")
```

Every column has a similar standard deviation and standard deviation of the mean, except column 5. There was an outlier in measurement #2 of column 5 which is causing these skewed numbers. This outlier was likely caused by a human error in setup. Other groups had similar data spreads. The standard deviations from my group are similar to Billhartz-Lopez, but low compared to Larner-LaMena.
