```python
import numpy as np
import matplotlib.pyplot as plt
from jin_devito import acoustic

import sys
# this should be changed to your own google drive path
```

```python
# let's start up a 2 layer acoustic model (e.g. wavefield in the water)
mod = acoustic.acoustic_model()  # create an instant of acoustic model class

x_range = 5000  # horizontal span of the model, in meter
z_range = 2000  # vertical span of the model, in meter
v1 = 1500        # seismic velocity of the first layer, in m/s
v2 = 2500        # seismic velocity of the second layer, in m/s
boundary_depth = 1000  # depth of the velocity boundary, in meter

# input these parameters to create a two layer model
mod.set_two_layer_model(x_range=x_range,
                        y_range = z_range,
                        v1=v1,
                        v2=v2,
                        boundary_depth=boundary_depth)

# set simulation time. This needs to be done before setting the source
total_simulation_time = 2000  # in ms
mod.set_model_time(total_simulation_time)

# set source location
src_x = 2500; src_z = 10;  # source location, in meters
mod.set_ricker_src(src_x,src_z)

# set receiver locations
receiver_number = 20       # number of receivers
rx = np.linspace(100,5000,receiver_number)  # receiver x coordinates
rz = np.ones_like(rx)*10     # receiver y coordinates
mod.set_receiver(rx,rz)
```

```python
# check model setup

# plot velocity model and receiver/source locations
# red dot is source, green triangles are receivers
mod.plot_model()
```

```python
# run the model to simulate acoustic wavefield

mod.run_model()
```

```python
# make a gif movie about the wavefield

gif_filename = './acoustic_wavefield.gif'

mod.make_wavefield_movie(gif_filename,plot_srcrec=True, writer='imagemagick')  # this will create a gif file in the current folder

# display the gif file we just generated
mod.IMAGE_display(gif_filename)
```
